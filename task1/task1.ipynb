```python
#importing required libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loading and reviewing data
data = pd.read_csv("sales_data.csv")
print(data)
```

```python
#line plot
plt.plot(data["ID"], data["Revenue"], c='r')
plt.title("ID vs Revenue line plot")
plt.xlabel("ID")
plt.ylabel("Revenue")
```

```python
#scatter plot
plt.scatter(data["Quantity"], data["Revenue"], c='r')
plt.title("Quantity vs Revenue scatter plot")
plt.xlabel("Quantity")
plt.ylabel("Revenue")
```

```python
#bar chart
tr_A = 0
tr_B = 0
tr_C = 0
for index, row in data.iterrows():
    if row["Product"]=='A':
        tr_A+=row["Revenue"]
    elif row["Product"]=='B':
        tr_B+=row["Revenue"]
    elif row["Product"]=='C':
        tr_C+=row["Revenue"]

total_revenue = [tr_A, tr_B, tr_C]
product = ['A', 'B', 'C']

plt.bar(product, total_revenue)
plt.title("Product vs Total Revenue bar plot")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
```
